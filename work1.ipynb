##pandasライブラリを用いたプログラム

pandasライブラリを用いて３教科のテストの点の平均点、分散、標準偏差を出力するプログラムを作成しました。

最初にcsvファイルを読み込むかランダムにデータを生成するかを選択することができます。

```python
import pandas as pd
import random

ans = input('３教科の点数が入力されているcsvファイルを読み込む場合は１、データをランダムに生成する場合は２を入力して下さい')
while ans != '2'and ans != '1':
  print('１または２だけ入力してください')
  ans = input('３教科の点数が入力されているcsvファイルを読み込む場合は１、データをランダムに生成する場合は２を入力して下さい')

if ans == '1':#３教科の点数が入力されているcsvファイルを読み込む
  x = input('csvファイル名を入力してください')
  df = pd.read_csv(x)

  df.columns = ['Japanese','Math','English']

  print('平均点\n国語：'+str(format(df['Japanese'].mean(),'.3f'))+'数学：'+str(format(df['Math'].mean(),'.3f'))+'英語：'+str(format(df['English'].mean(),'.3f')))
  print('分散\n国語：'+str(format(df['Japanese'].var(),'.3f')),'数学：'+str(format(df['Math'].var(),'.3f')),'英語：'+str(format(df['English'].var(),'.3f')))
  print('標準偏差\n国語：'+str(format(df['Japanese'].std(),'.3f')),'数学：'+str(format(df['Math'].std(),'.3f')),'英語：'+str(format(df['English'].std(),'.3f')))



if ans == '2': #データをランダムに生成する
  a = []
  n =input('何人分？')
  while True:
    try:
      n = int(n)

    except:
      print('自然数を入力してください')
      n = input('何人分？')
      continue

    if n <= 0:
      print('自然数を入力してください')
      n = input('何人分？')
    else :
      break


  for i in range(n):
    a.append([random.randint(0,100),random.randint(0,100),random.randint(0,100)])

  df = pd.DataFrame(a)
  df.columns = ['Japanese','Math','English']

  print('平均点\n国語：'+str(format(df['Japanese'].mean(),'.3f'))+'数学：'+str(format(df['Math'].mean(),'.3f'))+'英語：'+str(format(df['English'].mean(),'.3f')))
  print('分散\n国語：'+str(format(df['Japanese'].var(),'.3f')),'数学：'+str(format(df['Math'].var(),'.3f')),'英語：'+str(format(df['English'].var(),'.3f')))
  print('標準偏差\n国語：'+str(format(df['Japanese'].std(),'.3f')),'数学：'+str(format(df['Math'].std(),'.3f')),'英語：'+str(format(df['English'].std(),'.3f')))
df

```
